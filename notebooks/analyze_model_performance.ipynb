```python
import os
import json
import numpy as np
import pandas as pd
import SimpleITK as sitk
from tqdm import tqdm
from joblib import Parallel, delayed
from detection3d.utils.image_tools import resample_spacing

# Your global configs
target_image_origin = (0, 0, 0)
target_image_spacing = (0.5, 0.5, 0.5)  # Assuming isotropic spacing for simplicity
img_volume_path = '/home/mohanty/dental_output_new/image_nifti/'
target_dataset_path = "/home/mohanty/VSCODE_PROJECTS/Medical-Detection3d-Toolkit/detection3d/data/nasion_not_present_0.5"

os.makedirs(target_dataset_path, exist_ok=True)

def process_nifti_file(filename):
    try:
        image_path = os.path.join(img_volume_path, filename)
        patient_id = filename.replace('.nii', '')


        if not os.path.exists(image_path):
            return f"[SKIP] Image not found: {filename}"

        target_crop_volume_path = os.path.join(target_dataset_path, filename)

        # Crop Q1
        image_crop_Q1 = crop_cbct_quadrant(image_path, quadrant_name='Q1', return_image=True)
        old_origin = image_crop_Q1.GetOrigin()
        # Set origin
        image_crop_Q1.SetOrigin(target_image_origin)
        
        # Resample to target spacing
        if image_crop_Q1.GetSpacing() != target_image_spacing:
            image_crop_Q1 = resample_spacing(
                image_crop_Q1,
                target_image_spacing,
                1,
                'LINEAR'
            )

        spacing = image_crop_Q1.GetSpacing()
        
        if (image_crop_Q1.GetOrigin() == target_image_origin):
            sitk.WriteImage(image_crop_Q1, target_crop_volume_path, useCompression=True)
            # return f"[OK] {patient_id}"
        else:
            return f"[WARN] Invalid origin or negative landmarks: {patient_id}"

    except Exception as e:
        return f"[ERROR] {filename}: {str(e)}"

# Wrap crop_cbct_quadrant to return the image directly
def crop_cbct_quadrant(image_path, quadrant_name='Q1', return_image=False):
    image = sitk.ReadImage(image_path)
    size = image.GetSize()
    mid_y = size[1] // 2
    mid_z = size[2] // 2

    quadrants = {
        'Q1': ([0, 0, mid_z], [size[0], mid_y, size[2] - mid_z]),
        'Q2': ([0, mid_y, mid_z], [size[0], size[1] - mid_y, size[2] - mid_z]),
        'Q3': ([0, 0, 0], [size[0], mid_y, mid_z]),
        'Q4': ([0, mid_y, 0], [size[0], size[1] - mid_y, mid_z])
    }

    start, crop_size = quadrants[quadrant_name]
    cropped = sitk.RegionOfInterest(image, size=crop_size, index=start)

    if return_image:
        return cropped

    # fallback save-only mode (not used here)
    patient_id = os.path.splitext(os.path.basename(image_path))[0].replace('.nii', '')
    out_path = os.path.join(os.path.dirname(image_path), f"{patient_id}_{quadrant_name}.nii.gz")
    sitk.WriteImage(cropped, out_path)

# ✨ Run in parallel

results = Parallel(n_jobs=2)(delayed(process_nifti_file)(f) for f in tqdm(nasion_not_found['filename'], desc="Parallel Processing"))

# Optional: Print summary

# for r in results:
#     print(r)
```
